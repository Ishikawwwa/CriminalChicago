```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from pyspark.ml.feature import StringIndexer
from functools import reduce

# Convert Date to timestamp
crime_df = crime_df.withColumn("ts", F.to_timestamp("Date"))

# Create string version of District for encoding
crime_df = crime_df.withColumn("District_str", F.col("District").cast("string"))

# Encode categorical columns using StringIndexer
indexer_primary = StringIndexer(inputCol="Primary Type", outputCol="primary_type_enc")
indexer_fbi = StringIndexer(inputCol="FBI Code", outputCol="fbi_code_enc")
indexer_district = StringIndexer(inputCol="District_str", outputCol="district_enc")

crime_df = indexer_primary.fit(crime_df).transform(crime_df)
crime_df = indexer_fbi.fit(crime_df).transform(crime_df)
crime_df = indexer_district.fit(crime_df).transform(crime_df)

# Convert boolean columns to integer type
crime_df = crime_df.withColumn("Arrest", F.col("Arrest").cast("integer"))
crime_df = crime_df.withColumn("Domestic", F.col("Domestic").cast("integer"))

# Define base features for lagging
base_features = [
    'primary_type_enc', 'fbi_code_enc', 'Arrest',
    'Domestic', 'Beat', 'Ward', 'Community Area'
]

# Create window specification for temporal operations
window_spec = Window.partitionBy("district_enc").orderBy("ts")

# Calculate next timestamp and time difference label
crime_df = crime_df.withColumn("ts_next", F.lead("ts", 1).over(window_spec))
crime_df = crime_df.withColumn("label", 
    F.unix_timestamp("ts_next") - F.unix_timestamp("ts"))
crime_df = crime_df.filter(F.col("label").isNotNull())

# Generate lagged features for temporal patterns
lagged_columns = []
for feature in base_features:
    for lag in [1, 2, 3]:
        lag_col = f"{feature}_prev{lag}"
        crime_df = crime_df.withColumn(lag_col, F.lag(F.col(feature), lag).over(window_spec))
        lagged_columns.append(lag_col)

# Filter out rows with missing lagged values
filter_condition = reduce(
    lambda a, b: a & b, 
    (F.col(c).isNotNull() for c in lagged_columns), 
    F.lit(True)
)
crime_df = crime_df.filter(filter_condition)

# Prepare final feature set and label
feature_columns = base_features + ['district_enc'] + lagged_columns
X = crime_df.select(feature_columns)
y = crime_df.select("label")
```
